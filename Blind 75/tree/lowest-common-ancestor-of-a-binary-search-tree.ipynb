### Lowest Common Ancestor of a Binary Search Tree

```
medium
```

---

Given a binary search tree (BST), find the lowest common ancestor (LCA) of two given nodes in the BST.

According to the definition of LCA on Wikipedia: "The lowest common ancestor is defined between two nodes p and q as the lowest node in T that has both p and q as descendants (where we allow a node to be a descendant of itself)."

**Example 1:**

```
Input: root = [6,2,8,0,4,7,9,null,null,3,5], p = 2, q = 8
Output: 6
Explanation: The LCA of nodes 2 and 8 is 6.
```

Example 1 Tree Structure:

```
        6
       / \
      2   8
     /|   |\
    0 4   7 9
      /\
     3 5
```

**Example 2:**

```
Input: root = [6,2,8,0,4,7,9,null,null,3,5], p = 2, q = 4
Output: 2
Explanation: The LCA of nodes 2 and 4 is 2, since a node can be a descendant of itself according to the LCA definition.
```

Example 2 Tree Structure:

```
        6
       / \
      2   8
     /|   |\
    0 4   7 9
      /\
     3 5
```

**Example 3:**

```
Input: root = [2,1], p = 2, q = 1
Output: 2
```

Example 3 Tree Structure:

```
    2
   /
  1
```

**Constraints:**

- The number of nodes in the tree is in the range [2, 10^5].
- `-10^9 <= Node.val <= 10^9`
- All Node.val are unique.
- p != q
- p and q will exist in the BST.

### Solution

````python
from tree import TreeNode


class Solution:
    def lowestCommonAncestor(
        self, root: "TreeNode", p: "TreeNode", q: "TreeNode"
    ) -> "TreeNode":
        while True:
            if root.val < p.val and root.val < q.val:
                root = root.right
            elif root.val > p.val and root.val > q.val:
                root = root.left
            else:
                return root
````

### Test

````python
for root, p, q, result in [
    # root = [6,2,8,0,4,7,9,null,null,3,5], p = 2, q = 8 Output: 6
    (
        TreeNode(6, TreeNode(2, TreeNode(0), TreeNode(4, TreeNode(3), TreeNode(5))), TreeNode(8, TreeNode(7), TreeNode(9))),
        TreeNode(2, 0, TreeNode(4, TreeNode(3), TreeNode(5))),
        TreeNode(8, TreeNode(7), TreeNode(9)),
        TreeNode(6, TreeNode(2, TreeNode(0), TreeNode(4, TreeNode(3), TreeNode(5))), TreeNode(8, TreeNode(7), TreeNode(9))),
    ),
    # root = [6,2,8,0,4,7,9,null,null,3,5], p = 2, q = 4 Output: 2
    (
        TreeNode(6, TreeNode(2, TreeNode(0), TreeNode(4, TreeNode(3), TreeNode(5))), TreeNode(8, TreeNode(7), TreeNode(9))),
        TreeNode(2, TreeNode(0), TreeNode(4, TreeNode(3), TreeNode(5))),
        TreeNode(4, TreeNode(3), TreeNode(5)),
        TreeNode(2, TreeNode(0), TreeNode(4, TreeNode(3), TreeNode(5))),
    ),
    # root = [2,1], p = 2, q = 1 Output: 2
    (
        TreeNode(2, TreeNode(1)),
        TreeNode(2, TreeNode(1)),
        TreeNode(1),
        TreeNode(2, TreeNode(1)),
    ),
]:
    prediction = Solution().lowestCommonAncestor(root, p, q)
    assert str(prediction) == str(result), f"{prediction} != {result} for {root}, {p}, {q}"
````
